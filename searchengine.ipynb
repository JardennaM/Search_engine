```python
from elasticsearch import Elasticsearch
HOST = 'http://localhost:9200/'
es = Elasticsearch(hosts=[HOST])
!curl -X PUT "localhost:9200/all-the-news" 
```

```python
def insertArticle(index, doc_type, _id, article):
    json = {
        'id': _id,
        'title': article['title'],
        'publication': article['publication'],
        'author': article['author'] if type(article['author']) == str else '',
        'date': article['date'],
        'year': article['year'],
        'month': article['month'],
        'url': article['url'] if type(article['url']) == str else '',
        'content': article['content']
    }
    return es.index(index=index, doc_type=doc_type, id=_id, body=json)
```

```python
for i, row in data.iterrows():
    insertArticle('all-the-news', 'articles', i, row)
```

```python
res = es.search(index="all-the-news", body={"query" : {"term" : { "year" : 2016.0 }}})
print(res['hits'])
```
